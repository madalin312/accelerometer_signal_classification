```python
#reading and loading the data:
train_data_directory = "C:/Facultate/Master/An 1/Sem1/Practical Machine Learning/Project 1/train/train"
project_dir = "C:/Facultate/Master/An 1/Sem1/Practical Machine Learning/Project 1"
submission_train_data_directory = "C:/Facultate/Master/An 1/Sem1/Practical Machine Learning/Project 1/test/test"

#NOTE: not all of the signals have 150 samples
def load_train_input_data(train_data_path):
    x_data = []
    for csv, i in zip(os.listdir(train_data_path), range(len(os.listdir(train_data_path)))):
        signal = pd.read_csv(os.path.join(train_data_path, csv), header=None).to_numpy()
        
        if signal.shape[0] < 150:
            #alternative approach to handling incomplete signals: we fill the data with the mean for the signals that
            #have under 150 values:
            #x_mean = np.mean(signal[:,0])
            #y_mean = np.mean(signal[:,1])
            #z_mean = np.mean(signal[:,2])
            #auxiliary_sample = np.array([x_mean, y_mean, z_mean])
            
            #we make the last values of the signal stationary:
            auxiliary_sample = np.array([signal[-1,0], signal[-1,1], signal[-1,2]])
            for j in range(150-signal.shape[0]):
                signal = np.append(signal, [auxiliary_sample], axis=0)
        elif signal.shape[0] > 150:
            #eliminate the values over index 150 for the signals with more than 150 values:
            signal = signal[:-(signal.shape[0]-150)]
        x_data.append(signal)
    x_data = np.array(x_data)
    
    return x_data

def load_train_output_data():
    y_df = pd.read_csv(project_dir + "/train_labels.csv")
    return y_df
```

```python
def check_for_nulls(x_data):
    nullValues = 0;
    for k in range(len(x_data)):
        for j in range(x_data[k].shape[1]):
            for i in range(x_data[k].shape[0]):
                if np.isnan(x_data[k][i,j]):
                    nullValues+=1
    if nullValues > 0:
        print("There are null values in the signals")
    else:
        print("There are no null values in the signals")
        
check_for_nulls(x_data)
```

```python
#we shift the values by 1 to the left so the labels start from 0. This is necessary for training the neural network afterwards
def encode_labels(y_df, y_data):
    labels = y_df["class"].unique()
    labels = np.sort(labels)
    encoding = {}
    for label in labels:
        encoding[label] = label-1
    for i in range(len(y_data)):
        y_data[i] = encoding[y_data[i]]
    return encoding, y_data
```

```python
encoding, y_data = encode_labels(y_df, y_data)
#y_data
```

```python
#creating the network:
def compile_network():
    dl_model = keras.models.Sequential()
    dl_model.add(tf.keras.layers.Flatten(input_shape=(150, 3)))
    dl_model.add(keras.layers.Dense(units=192, kernel_initializer="uniform", activation="relu"))
    dl_model.add(keras.layers.Dense(units=352, kernel_initializer="uniform", activation="relu"))
    dl_model.add(keras.layers.Dense(units=64, kernel_initializer="uniform", activation="relu"))
    dl_model.add(keras.layers.Dense(units=288, kernel_initializer="uniform", activation="relu"))    
    dl_model.add(keras.layers.Dropout(0.2))
    
    dl_model.add(keras.layers.Dense(20, kernel_initializer='uniform', activation='softmax'))
    optimizer = Adam(learning_rate=0.001) 
    dl_model.compile(
        optimizer = optimizer,
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy'])
    
    return dl_model

def compile_conv_dl_model():
    model = keras.Sequential()
    kernel_size = 3
    model.add(keras.layers.Conv2D(filters = 96,
                                    kernel_size = (kernel_size,kernel_size),
                                    padding = 'Same',
                                    activation = "relu",
                                    input_shape = (150,3,1)))
    model.add(keras.layers.Conv2D(filters = 72,
                                    kernel_size = (kernel_size,kernel_size),
                                    padding = 'Same',
                                    activation = "relu",
                                    input_shape = (150,3,1)))
    
    kernel_size = 5
    model.add(keras.layers.Conv2D(filters = 48,
                                    kernel_size = (kernel_size,kernel_size),
                                    padding = 'Same',
                                    activation = "relu",
                                    input_shape = (150,3,1)))
    model.add(keras.layers.Conv2D(filters = 112,
                                    kernel_size = (kernel_size,kernel_size),
                                    padding = 'Same',
                                    activation = "relu",
                                    input_shape = (150,3,1)))
    
    model.add(keras.layers.MaxPool2D(pool_size=(2,2), padding="Same"))
    model.add(keras.layers.Dropout(0.25))
        
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(256, activation = "relu"))
    model.add(keras.layers.Dropout(0.25))
    model.add(keras.layers.Dense(20, kernel_initializer='uniform', activation = "softmax"))
    optimizer = Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999)
    
    model.compile(
        optimizer = optimizer,
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy'])
    return model
```

```python
random_seed = 19
random.seed(random_seed)
def train_dl_model():
    #dl_model = compile_network()
    dl_model = compile_conv_dl_model()
    #training the network:
    kfold = KFold(n_splits = 5, shuffle = True, random_state=random_seed)
    epochs = 10
    for train_index, test_index in kfold.split(x_data):
        X_train, X_test, Y_train, Y_test = x_data[train_index], x_data[test_index], y_data[train_index], y_data[test_index]
        dl_model.fit(X_train, Y_train, epochs = epochs)
            
    return dl_model
```

```python
def save_nn_model(model, model_name):
    model_json = model.to_json()
    with open(model_name + ".json", "w") as json_file:
        json_file.write(model_json)

    # serialize weights to HDF5
    model.save_weights(model_name + ".h5")
    return


def load_nn_model(model_file_name):
    # load json and create model
    json_file = open(model_file_name + ".json", 'r')
    loaded_model_json = json_file.read()
    json_file.close()
    loaded_model = model_from_json(loaded_model_json)
    # load weights into new model
    loaded_model.load_weights(model_file_name + ".h5")
    print("Loaded model from disk")
    return loaded_model
    
```

```python
#functions used for hyperparameter tuning:
def build_dl_model(hp):
    model = keras.Sequential()
    model.add(tf.keras.layers.Flatten(input_shape=(150, 3)))
    for i in range(hp.Int('num_layers', 2, 25)):
        model.add(keras.layers.Dense(units = hp.Int('units' + str(i), min_value=32, max_value=512, step=32),
                               kernel_initializer= hp.Choice('initializer', ['uniform', 'normal']),
                               activation= hp.Choice('activation', ['relu', 'sigmoid', 'tanh'])))
    model.add(keras.layers.Dropout(0.2))
    model.add(keras.layers.Dense(20, kernel_initializer= hp.Choice('initializer', ['uniform', 'normal']), activation='softmax'))
    optimizer = Adam(learning_rate=0.001) 
    model.compile(
        optimizer = optimizer,
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy'])
    return model

def build_conv_net_model(hp):
    model = keras.Sequential()
    for i in range(hp.Int('num_layers', 2, 5)):
        kernel_size = hp.Int('kernel_size' + str(i), min_value=3, max_value=7, step=2)
        model.add(keras.layers.Conv2D(filters = hp.Int('filters' + str(i) + str(0), min_value=8, max_value=128, step=8),
                                      kernel_size = (kernel_size,kernel_size),
                                      padding = 'Same',
                                      activation = "relu",
                                      input_shape = (150,3,1)))
        model.add(keras.layers.Conv2D(filters = hp.Int('filters' + str(i) + str(1), min_value=8, max_value=128, step=8),
                                      kernel_size = (kernel_size,kernel_size),
                                      padding = 'Same',
                                      activation = "relu",
                                      input_shape = (150,3,1)))
        model.add(keras.layers.MaxPool2D(pool_size=(2,2), padding="Same"))
        model.add(keras.layers.Dropout(0.25))
        
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(256, activation = "relu"))
    model.add(keras.layers.Dropout(0.25))
    model.add(keras.layers.Dense(20, kernel_initializer='uniform', activation = "softmax"))
    optimizer = Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999)
    
    model.compile(
        optimizer = optimizer,
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy'])
    return model
```

```python
# hyperparameter tuning using random search:
tuner = kt.RandomSearch(
    build_dl_model,
    objective='val_accuracy',
    executions_per_trial=3,
    directory='project_knn_part', project_name = 'Smartphone User Identification - neural network 1')

tuner.search_space_summary()
```

```python
random_seed = 16
random.seed(random_seed)

#function used for hyperparameter tuning:
def train_model(model, grid):
    kfold = KFold(n_splits = 5, shuffle = True, random_state=random_seed)
    grid_search = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, scoring='accuracy',error_score=0)
    
    results = []
    for train_index, test_index in kfold.split(x_flattened):
        X_train, X_test, Y_train, Y_test = x_flattened[train_index], x_flattened[test_index], y_data[train_index], y_data[test_index]
        grid_result = grid_search.fit(X_train, Y_train)
        print("Best performance: %f, Parameters: %s" % (grid_result.best_score_, grid_result.best_params_))
        results.append(grid_result.best_score_)
    return results
```
